```python
import pandas as pd
```

```python
df = pd.read_csv('data/global-superstore-data.csv', encoding="utf-8")
```

```python
df.info()
```

```python
df.columns
```

```python
dim_Customer = ['Customer ID', 'Customer Name', 'Segment']
dim_Product = ['Product ID', 'Category','Sub-Category', 'Product Name']
dim_Geography =['Address ID', 'City', 'State', 'Country','Postal Code', 'Market', 'Region']
dim_Order = ['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Order Priority']

fact_Sale = ['Order ID', 'Product ID', 'Customer ID', 'Address ID', 'Sales', 'Quantity', 'Discount','Profit', 'Shipping Cost']
```

```python
Address_df = df[['City', 'State', 'Country','Postal Code', 'Market', 'Region']]
Address_df = Address_df.drop_duplicates()
Address_df['Address ID'] = Address_df.index+1
Address_df
```

```python
dff = pd.merge(df, Address_df, on = ['City', 'State', 'Country','Postal Code', 'Market', 'Region'], how = 'outer')
```

```python
dff = dff.set_index('Row ID')
```

```python
dim_Geography_df = dff[dim_Geography]
dim_Geography_df = dim_Geography_df.drop_duplicates()
dim_Geography_df
```

```python
dim_Customer_df = dff[dim_Customer]
dim_Customer_df = dim_Customer_df.drop_duplicates()
dim_Customer_df
```

```python
dim_Product_df = dff[dim_Product]
dim_Product_df = dim_Product_df.drop_duplicates()
dim_Product_df = dim_Product_df.drop_duplicates(subset='Product ID', keep='first')
dim_Product_df
```

```python
dim_Order_df = dff[dim_Order]
dim_Order_df = dim_Order_df.drop_duplicates()
dim_Order_df = dim_Order_df.drop_duplicates(subset='Order ID', keep='first')

dim_Order_df
```

```python
fact_Sale_df = dff[fact_Sale]
fact_Sale_df = fact_Sale_df.drop_duplicates()
fact_Sale_df
```

```python
duplicate_ids = dim_Product_df[dim_Product_df['Product ID'].duplicated(keep=False)]

print(duplicate_ids)
```

```python
dim_Customer_df.to_csv('data/tables/Customer.csv', encoding="utf-8", index=False)
dim_Geography_df.to_csv('data/tables/Geography.csv', encoding="utf-8",index=False)
dim_Product_df.to_csv('data/tables/Product.csv', encoding="utf-8",index=False)
dim_Order_df.to_csv('data/tables/Order.csv',encoding="utf-8", index=False)
fact_Sale_df.to_csv('data/tables/Sales.csv',encoding="utf-8", index=False)
```
